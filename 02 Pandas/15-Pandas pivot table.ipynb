## Pandas pivot table 

```python
# Pandas Pivot table --------  The levels in the pivot table will be stored in multindex objects (hierarchical indexes) on 
#                              the index and columns of the result DataFrame 
```

```python
# pivot table ka use  karne se jo result data milta hai usme pivot table ke levels multiindex object me store hota hai 
# pivot table ka use hum DataFrame ke all data ko sumarize karne ke liye karte hai aur ye return karta hai new data frame 
```

```python
import pandas as  pd
```

```python
# import the data set 
df1 = pd.read_csv("C:\\Users\\Admin\\Documents\\Movie  Rateings.csv")
```

```python
df1
```

 #### Now apply pivot function

```python
# pivot table ka use karke hum dataset ko sumarize kar sakte hai 
```

#### Index parameter 

```python
df1.pivot_table(index ='Genre')

# Here the data will be sumarize same type of data is summarize  
```

####  Column parameter 

```python
df1.pivot_table(index ='Genre',columns = 'Year')

# pahele humko columns ke according data ki inforation mil rahi thi
# ab humko year wais data ki information mil rahi hai 
```

#### aggfunc parameter

```python
df1.pivot_table(index ='Genre',columns = 'Year',aggfunc  = 'count')
```

```python
df1.pivot_table(index ='Genre',columns = 'Year',aggfunc  = 'max')
```

```python
df1.pivot_table(index ='Genre',columns = 'Year',aggfunc  = 'sum')
```

### fill_value

```python
# hum pivot table me NaN value ki janghaa koi bhi value ko fill kar sakte hai 
# uske liye hum Fill_vallue parameter ka use karenge 
```

```python
df1.pivot_table(index ='Genre',columns = 'Year',aggfunc  = 'sum',fill_value = 'Not Defined')
```

### margins function

```python
# is function ak use karne ke baad humko yek all name ka columns milega jo values ko total karke show karta hai 
```

```python
df1.pivot_table(index ='Genre',columns = 'Year',fill_value = 'Not Define',margins = True)

# all column me average value print hoga 
```

```python
# hum yaha jo bhi function use karenge uska mean value humko milega like hum yaha sum lenge to uska mean value milega

df1.pivot_table(index ='Genre',columns = 'Year',aggfunc = 'sum',fill_value = 'Not Define',margins = True)
```
