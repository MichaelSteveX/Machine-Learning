# nrows 

This function is use for particular rows check 

```python
# if humko pure data me se kisi particular row  ko find out karna hai to hum is function ka use karenge 
# usi dataset ke andar hum function ko  addd karenge 

df = pd.read_csv("/kaggle/input/fortune-data-set/Fortune_10.csv",nrows = 4)
df

# jitne rows ka input denge hum utne rows humko display hoga 
```

# usecols

This function is use to check particular column 


```python
# if humko pure data me se kisi particular columns ko find out karna hai to hum is function ka use karenge 
# usi dataset ke andar hum function ko  addd karenge

df = pd.read_csv("/kaggle/input/fortune-data-set/Fortune_10.csv",usecols =[0])
df
```

```python
# if we want to print first and second column 
df = pd.read_csv("/kaggle/input/fortune-data-set/Fortune_10.csv",usecols =[0,1,2])
df
```

# skiprows

This function is use to skip the particular row from the table 

```python
# if humko kisi  row ko skip karna hai to hum ye function ka use karenge 

df = pd.read_csv("/kaggle/input/fortune-data-set/Fortune_10.csv",skiprows = 1)
df

# agar hum skip rows ka use karnge to wo particular rows hat jayega 
```

```python
# if humko 2 rows ko skip karna hai to uske liye hum innput change karenge 

df = pd.read_csv("/kaggle/input/fortune-data-set/Fortune_10.csv",skiprows = [5])
df
```

```python
# hum isme multiple rows ko bhi skip kar sakte hai 
# sabhi rows ka humko index dena padega list me 

df = pd.read_csv("/kaggle/input/fortune-data-set/Fortune_10.csv",skiprows = [1,2,3])
df
```

```python
# agar humko kisis paricular column ko chnage karke usko index banana hai to hum is function ka use karenge

df = pd.read_csv("/kaggle/input/fortune-data-set/Fortune_10.csv")
df
```

```python
# ye hamara pura data hai aur iske id ko mujhe index banana hai to hum ye function ka use karenge 

df = pd.read_csv("/kaggle/input/fortune-data-set/Fortune_10.csv",index_col = "ID")
df

# To isme hamara ID index ban jayega 
```

```python
# if humko name ko index banana hai to hum function ko change karenge 

df = pd.read_csv("/kaggle/input/fortune-data-set/Fortune_10.csv",index_col = "Name")
df

```
