# Data Frame with Dictionary 

yaha par humne dictionary ke andar list use kiya hai with the help of list humne data frame banaya hai 


```python
dict1 = {'ID': [1,2,3,4,5,6,7,8]}
```

```python
# single column data frame 

dict2 = pd.DataFrame(dict1)
dict2
```

```python
# multiple column data frame 


dict1 = {'ID': [1,2,3,4,5,6,7,8],'NAME': ['Suresh','Ramesh','Suraj','Ateek','Suraj','Ravi','Aman','Yusuf']}

# fist {[]} dict fir uske andar list 
```

```python
dict2 = pd.DataFrame(dict1)
dict2
```

# List of Dict  

yaha par hum list ke andar dict ko create karenge fir data frame banayenge 

```python
ls_dict = [{'A': 1,'B':2,'C':3},{'A':3,'B':5,'C':7}]
```

```python
dict3 = pd.DataFrame(ls_dict)
dict3

# yaha pr humne list ke andar dict ka use karke data frame ko create kar liya hai 
```

```python
# we made a data frame with unequal values 

ls_dict = [{'A': 1,'B':2,'C':3,'D':34},{'A':3,'B':5,'C':7,},{'A':2,'B':3,'C':34,}]
```

```python
dict4 = pd.DataFrame(ls_dict)
dict4

# in the pandas if the value is not given then it can take a null value automatically 
```

# Dict of Series 

Means dictionary ke andar series ka use karke hum data frame ko create kar sakte hai 

```python
dict_sr = {'id': pd.Series([1,2,3,4]),'SN':([111,222,333,444])}
dict6 = pd.DataFrame(dict_sr)
dict6

# yaha humne dict ke andar series ka use karke data frame ko create kar liya 
```
