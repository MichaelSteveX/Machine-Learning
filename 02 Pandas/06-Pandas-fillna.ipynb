# Pandas fillna

```python
# pandas fillna method means in the dataset where the NaN is given then this method replace the NaN with new values 
```

```python
import pandas as pd
```

```python
# iport data 

df = pd.read_csv('/kaggle/input/fortune-data-set/Fortune_10.csv')
df
```

```python
# in the table jaha par nan show kar raha hai waha par hum fillna method ka use karke usko fill karenge 

df.fillna('3,344,546 Dollars')

# the NaN is remove and given value is printed
```

```python
# agar humko har yek specific column ko yek specific value dena hai nan value ki jangha to hum dictionary ka use karenge 
df.fillna({'Expenses':'paid'})


# yaha par column ke andar humari given value print ho gai 
```

```python
# is method ka use hum tab karte hai jab hum nan ki value pervious index se he fill karna ho to 

df.fillna(method = 'ffill')  # forward 

# to yaha par nan value me previos value fill ho chuki hai in row 9 upper ko copy karke lower me paste kar deta hai 
```

```python
df.fillna(method = 'bfill') # for backword
```

```python
# limit function ka tab hum use karte hai jab humko specific nan value ko print karna hota hai 
# isme value and method dono ko specifie karna hai 
df.fillna(0, limit = 1)
```
