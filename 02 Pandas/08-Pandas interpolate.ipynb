#### Axis parameter

```python
# hum yaha par missing values ko fill karenge axis ke accordind like if we take 0 then thats mens is 'column' and if we take 1 that mens is 'row'
```

```python
df = pd.read_csv('C:\\Users\\Admin\\Downloads\\Student data.csv')
df # we call the dataset 
```

```python
# is function ko execute karne ke liye hamara data type object nahi hona chaiye 
# tp iske liye humko check  karna padega 
df.dtypes
```

```python
df.interpolate(axis = 0) # zero mens column by column and 1 mens row by row 
```

#### limit parameter 

```python
# limit parameter ka use hhum tab karte hai jab humko particular number me missing values ko fill karna ho 
```

```python
df.interpolate(limit = 1)

# if hum default 1 value denge to yaha sirf one 1 row he fill hoga if 2 denge to 2 rows fill hoga 
```

```python
# Now take limit = 2
```

```python
df.interpolate(limit = 2)
```

#### limit_direction

```python
# limit direction mens hum apne values ko forward , backword dono form me fill kar sakte hai
```

```python
df.interpolate(limit = 1,limit_direction = 'backward')
```

```python
df.interpolate(limit = 1,limit_direction = 'both')
```

```python
# inplace parameter boolian formate me input late hai 
df.interpolate(inplace = True)
df
```
