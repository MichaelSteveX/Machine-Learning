## Pandas concat 

```python
# pandas concat function --- pandas provides various facilities for easily combinig together series
#                           Data frame and panel object 

# series , DataFrame and Panel ko appas me combine karne ke liye hum pandas concat function ka use karte hai 

# Basically concat ka matlab hota hai yek data ko dusre data ke uppar set karna like staking  

```

```python
import pandas as pd 
```

#### Concat with series

```python
# Now we have to create a data 
sr1 = pd.Series([0,1,2,3])
sr1
```

```python
sr2 = pd.Series([3,4,5,6,7,8])
sr2
```

```python
# Here two series are created now we use the concat function 
```

##### Now we use Concat and their parameter

#### objs

```python
pd.concat([sr1,sr2])

# here both the seires are concat with column wise 
# this is 2 equal element series 
```

```python
# now 2 unequaal series concat 
pd.concat([sr1,sr2])
```

#### Concat with DataFrame

```python
# first data frame 

df1 = pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','c','D'],
             'Class':[5,6,7,8]})
df1
```

```python
# second data frame
df2 = pd.DataFrame({'ID':[1,2,3,4],'Name':['E','F','G','H'],
             'Class':[9,10,11,12]})
df2
```

```python
# Now cancat both the data frame 
pd.concat([df1,df2])

# here both data is combined df2 
```

```python
# if we want to change the patern of dataframe 
# like data frame 2 first and data frame 1 in last 
```

```python
pd.concat([df2,df1]) # so we change the pattern 
```

#### Axis parameter 

```python
# concat dataframe column wise 
pd.concat([df1,df2],axis = 1)
```

```python
# concat dataframe Row wise 
pd.concat([df1,df2],axis = 0)
```

```python
# if index ko humko systematic way me rakhna hai to hum parameter denge 
pd.concat([df1,df2],axis = 0,ignore_index = True)

# isko run karne ke baadd hamara index systematic ho jayega 
```

```python
# we also cocate unequal dataframe
```

```python
 
df3= pd.DataFrame({'ID':[1,2,3,4],'Name':['E','F','G','H'],
             'Class':[9,10,11,12]})
df3
```

```python
df4 = pd.DataFrame({'ID':[1,2],'Name':['E','F'],
             'Class':[9,10]})
df4
```

```python
pd.concat([df3,df4]) # this concat is row wise 
```

```python
pd.concat([df3,df4],axis = 1) # this concat is column wise 
```

#### join parameter 

```python
# if we want to print only common value from both the data frame then we use join parameter 
```

```python
pd.concat([df3,df4],axis = 1,join = 'inner') # it gives only common values 
```

#### Keys parameter 

```python
# key parameter mens hum kisi bhi data ko yek label de sakte hai 
```

```python
# Call the data frame 
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','c','D'],
             'Class':[5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,3,4],'Name':['E','F','G','H'],
             'Class':[9,10,11,12]})
df2
```

```python
# if we want to seperate  dataframe with specific row then we use this function 
```

```python
pd.concat([df1,df2] , keys = ['df1','df2'])
```

```python
pd.concat([df1,df2] , keys = ['First','Second'])
```

```python
# axis parameter 

pd.concat([df1,df2] , keys = ['First','Second'],axis = 1)
```
