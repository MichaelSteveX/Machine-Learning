## Pandas melt 

```python
# pandas melt function -------pandas melt function is used to transform or reshape data 
#  mens pandas melt function ka use data ko transform and reshape karneke liye hota hai 
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv("C:\\Users\\Admin\\Documents\\Movie  Rateings.csv")
df1
```

#### Melt function 

```python
pd.melt(df1)
```

```python
pd.melt(df1,id_vars = ['Genre'])
```

```python
pd.melt(df1,id_vars = ['Year'])
```

```python
# if we want single value according to the year then we use this parameter 
pd.melt(df1,id_vars = ['Year'],value_vars = ['Genre'])

# yaha pe jo value mili hai wo according to the genre mila hai 
```

```python
pd.melt(df1,id_vars = ['Year'],value_vars = ['Ratings%'])

# to yaha par sirf humko ratings ka data show hoga 
```

```python
# if we want to change the header name of the column then we use this function 

pd.melt(df1,id_vars = ['Year'],value_vars = ['Ratings%'],var_name = 'Category')
```

```python
# if we want to chnage the headdder name of the value then we use this  function 
pd.melt(df1,id_vars = ['Year'],value_vars = ['Genre'],var_name = 'Category',value_name = 'Data')
```
