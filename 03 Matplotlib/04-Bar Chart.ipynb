# Bar chart

### Ploting Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
# Data set of 'Indian Artificial Intelligence Production (IAIP) Class'

classes = ['Python','R','Ai','ML','DS']
class1_student = [30,10,20,25,10]
class2_student = [40,5,20,20,10]
class3_student = [35,5,30,15,15]
```

#### Bar graph

```python
# plot histogram 
plt.bar(classes,class1_student)
```

```python
# graph showing horizontal(barh) 
plt.barh(classes,class1_student)
```

#### Parameter of Bar Chart

##### Width 

```python
# Giving a width parameter to the bar chart 
plt.bar(classes,class1_student,width = 0.2)
```

#### align 

```python
# align perameter hamara by default center hota hai 
# hum graph ko apne according align bhi kar sakte hai 
plt.bar(classes,class1_student,width = 0.2,align = 'edge')

# edge function use karne ke baad sare bars edge par aagaye 
```

#### Color

```python
# if we want to change the color of bar  and edge 
plt.bar(classes,class1_student,width = 0.2,align ='center',color = 'y',edgecolor = 'r' )
```

#### line width 

```python
# if we want to change the width of the bar line then we use this parameter 
plt.bar(classes,class1_student,width = 0.2,align ='center',color = 'y',edgecolor = 'r',
       linewidth = 4)
```

#### Alpha

```python
# if we want to increase or decrees the transperancy of the graph then we use this parameter 
plt.bar(classes,class1_student,width = 0.2,align ='center',color = 'y',edgecolor = 'r',
       linewidth = 2,alpha = 0.5)
```

#### linestyle

```python
# if we want to change the line of the bar then we use this parameter 
plt.bar(classes,class1_student,width = 0.2,align ='center',color = 'y',edgecolor = 'r',
       linewidth = 2,alpha = 0.5,linestyle = '--')
```

#### label 

```python
# if we want to give a lable of the particular graph then we use this parameter 
plt.bar(classes,class1_student,width = 0.2,align ='center',color = 'y',edgecolor = 'r',
       linewidth = 2,alpha = 0.5,linestyle = '--',label = 'Student')
plt.legend(loc = 1)
```

#### Visible 

```python
# visible parameter is used to hide the graph 
plt.bar(classes,class1_student,width = 0.2,align ='center',color = 'y',edgecolor = 'r',
       linewidth = 2,alpha = 0.5,linestyle = '--',label = 'Student',visible = False)
plt.legend(loc = 1)
```

## Bar chat concat

```python
# in this we have a multiple dataset and we have to plot all the dataset in to single bar chart 
```

```python
# Now change the background 
style.use('ggplot')
plt.bar(classes,class1_student,width = 0.2,align ='center',color = 'y',edgecolor = 'k',
       linewidth = 2,alpha = 0.5,linestyle = '-',label = 'Student')
plt.grid(color='w', linestyle='-', linewidth=2)
plt.legend(loc = 1)
```

#### figure

```python
# figure finction ka use karke hum chart ko apne according ardest kar sakte hai 
style.use('ggplot')
plt.figure(figsize = (8,5))
plt.bar(classes,class1_student,width = 0.2,align ='center',color = 'y',edgecolor = 'k',
       linewidth = 2,alpha = 0.5,linestyle = '-',label = 'Student')
plt.legend(loc = 1)
```

#### Horizontal

```python
# if we want a graph in horizontal then we use barh
style.use('ggplot')
plt.figure(figsize = (8,5))
plt.barh(classes,class1_student,align ='center',color = 'y',edgecolor = 'k',
       linewidth = 2,alpha = 0.5,linestyle = '-',label = 'Student')
plt.legend(loc = 1)
```

#### Title and x and y label


```python
style.use('ggplot')
plt.figure(figsize = (8,5))
plt.bar(classes,class1_student,align ='center',color = 'y',edgecolor = 'k',
       linewidth = 2,alpha = 0.5, width = 0.5,linestyle = '-',label = 'Student')
plt.title('Programming Language',fontsize = 15)
plt.xlabel('No Of Students',fontsize = 12)
plt.ylabel('Language',fontsize = 12)
plt.legend(loc = 1)
plt.show()
```

#### Showing multiple  data in single graphs

```python
# if you want to show 2 or 3 data in same graph then we do this step 
style.use('ggplot')
plt.figure(figsize = (8,5))


plt.bar(classes,class1_student,align ='center',color = 'k', width = 0.2,label = 'Student_1')

plt.bar(classes,class2_student,align ='center',color = 'y', width = 0.4,label = 'Student_2')

plt.bar(classes,class3_student,align ='center',color = 'm',width = 0.6,label = 'Student_3')




plt.title('Programming Language',fontsize = 15)
plt.xlabel('No Of Students',fontsize = 12)
plt.ylabel('Language',fontsize = 12)
plt.legend(loc = 1)
plt.show()

# if we use this methos then every graph is overlaped each other which is not requird we wont splitted lines graph 
```

#### if three data ko yek he graph pe he show karna hai to hum is type ke parameter ka use karenge 

```python
# so we plot a grph with having a both data without any overlapping 
# so we using a sum different parameters for this graph representation 

classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index, class1_student,width, color = 'k',alpha = 0.5,label = 'Student_1')

plt.bar(classes_index + width,class2_student,width, color = 'y',alpha = 0.5,label = 'Student_2')

plt.bar(classes_index + width + width,class3_student,width, color = 'm',alpha = 0.5,label = 'Student_3')
plt.xticks(classes_index + width ,classes) # ye function bar graph ki naming ke liye use hota hai 
plt.title('Programming Language',fontsize = 15)
plt.xlabel('No Of Students',fontsize = 12)
plt.ylabel('Language',fontsize = 12)
plt.legend(loc = 1)
plt.show()
```

#### Horizontal bar plot 

```python
classes_index = np.arange(len(classes))
width = 0.2
plt.barh(classes_index, class1_student,width, color = 'k',alpha = 0.5,label = 'Student_1')

plt.barh(classes_index + width,class2_student,width, color = 'y',alpha = 0.5,label = 'Student_2')

plt.barh(classes_index + width + width,class3_student,width, color = 'm',alpha = 0.5,label = 'Student_3')
plt.yticks(classes_index + width ,classes) # ye function bar graph ki naming ke liye use hota hai 
plt.title('Programming Language',fontsize = 15)
plt.xlabel('No Of Students',fontsize = 12)
plt.ylabel('Language',fontsize = 12)
plt.legend(loc = 1)
plt.show()
```
