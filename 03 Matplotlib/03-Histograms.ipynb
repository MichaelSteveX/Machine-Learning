```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Now generate the data of two student 
# Now we use here the numpy rendom modelto generate the rendom dataply
Machine_learning_student_age = np.random.randint(18,45,(100))
Python_learning_student_age =  np.random.randint(15,40,(100))
```

```python
print(Machine_learning_student_age)
print(Python_learning_student_age )
```

### Histograms 

```python
plt.hist(Machine_learning_student_age)
plt.show

# ye hammara normal show graph hai ab isme hum modification karenege 
```

```python
# first isme hum title add karenege 
plt.hist(Machine_learning_student_age)
plt.title("ML Student Age Histograms")
plt.show
```

```python
# second isme hum label denge x and y ko 
plt.hist(Machine_learning_student_age)
plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.show
```

# Hits parameter 

## bins

```python
# now we modify the graph with the hits parameter 
bins = [15,20,25,30,35,40,45]
plt.hist(Machine_learning_student_age,bins)
plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.show
```

```python
# if bar ki width ko humko km karna hai to hum rwidth ka use karenge 
bins = [15,20,25,30,35,40,45]
plt.hist(Machine_learning_student_age,bins,rwidth = 0.7)
plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.show
```

## histtype

```python
# histtype ka ue karke hum graph ka outline check kar sakte hai 
bins = [15,20,25,30,35,40,45]
plt.hist(Machine_learning_student_age,bins,rwidth = 0.7,histtype = 'bar',align = 'left')
plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.show
```

## orientation

```python
# orientation ka use karke hum graph ko change kar sakte hai like vertical to horizontal and horizontal to vertical 
bins = [15,20,25,30,35,40,45]
plt.hist(Machine_learning_student_age,bins,rwidth = 0.7,histtype = 'bar',orientation = 'horizontal')
plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.show
```

## color

```python
# if bar ko  humko color dena hai to hum color parameter ka use karenege 
bins = [15,20,25,30,35,40,45]
plt.hist(Machine_learning_student_age,bins,rwidth = 0.7,histtype = 'bar',orientation = 'vertical',color = 'm')
plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.show
```

## label 

```python
# histograms ko hum lebel kar sakte hai 
# label  function ka use karne ke liye humko "leegend" function ka  use karna padega 
bins = [15,20,25,30,35,40,45]
plt.hist(Machine_learning_student_age,bins,rwidth = 0.7,histtype = 'bar',orientation = 'vertical',color = 'm',label = 'Ml Student')
plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.legend(loc = 2)
plt.show
```

```python
# yek he histogram me do student ke data ko show karrna hai 
bins = [15,20,25,30,35,40,45]

plt.hist(Machine_learning_student_age,bins,rwidth = 0.7,histtype = 'bar',orientation = 'vertical',color = 'm',label = 'Ml Student')
plt.hist(Python_learning_student_age,bins,rwidth = 0.5,histtype = 'bar',orientation = 'vertical',color = 'y',label = 'Py Student')

plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.legend(loc = 2)
plt.show

# ye hamara dono graph yek he pe plot hua hai to isko hum alag alag plot karenge 
```

```python
# graph ko humko alag alag plot karna hai
bins = [15,20,25,30,35,40,45]

plt.hist([Machine_learning_student_age,Python_learning_student_age],bins,rwidth = 0.7,histtype = 'bar',orientation = 'vertical',color = ['m','y'],label = ['Ml Student','Py Student'])
# plt.hist(Python_learning_student_age,bins,rwidth = 0.5,histtype = 'bar',orientation = 'vertical',color = 'y',label = 'Py Student')

plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.legend(loc = 2)
plt.show
```

## Figure

```python
# figure function ka use hum histogram ki size ko badhaney ke liye karte hai 
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (7,7)) # isko use karke  hum histogram ki width and height chnage kar sakte hai 
plt.hist([Machine_learning_student_age,Python_learning_student_age],bins,rwidth = 0.7,histtype = 'bar',orientation = 'vertical',color = ['m','y'],label = ['Ml Student','Py Student'])
# plt.hist(Python_learning_student_age,bins,rwidth = 0.5,histtype = 'bar',orientation = 'vertical',color = 'y',label = 'Py Student')

plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.legend(loc = 2)
plt.show
```

## Background change 

```python
from matplotlib import style
style.use('ggplot')
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (10,4)) # isko use karke  hum histogram ki width and height chnage kar sakte hai 
plt.hist([Machine_learning_student_age,Python_learning_student_age],bins,rwidth = 0.7,histtype = 'bar',orientation = 'vertical',color = ['m','y'],label = ['Ml Student','Py Student'])
# plt.hist(Python_learning_student_age,bins,rwidth = 0.5,histtype = 'bar',orientation = 'vertical',color = 'y',label = 'Py Student')

plt.title("ML Student Age Histograms")
plt.xlabel('Student Age')
plt.ylabel('Number of Students')
plt.legend(loc = 2)
plt.grid(color='w', linestyle='-', linewidth=1)
plt.show
```
