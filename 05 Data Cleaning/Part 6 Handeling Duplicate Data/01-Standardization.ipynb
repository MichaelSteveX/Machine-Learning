## Standardization 

```python
# import requird library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# import the dataset
dataset = pd.read_csv('loan.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
# now find all the numerical column 
dataset.select_dtypes(include='float').columns
```

```python
# now fill the missing value 
for i in dataset.select_dtypes(include = 'float64').columns:
    dataset[i].fillna(dataset[i].mean(),inplace = True)
```

```python
dataset.isnull().sum()
```

```python
# nature of datase
sns.distplot(dataset['ApplicantIncome'])
plt.show()
```

```python
dataset.describe()
```

## Scalling with Scikit Learn 

```python
# Scalling with the help of scikit learn 
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
std.fit(dataset[['ApplicantIncome']])
```

```python
dataset['ApplicantIncome_ss'] = pd.DataFrame(std.transform((dataset[['ApplicantIncome']])),columns = ['X'])
```

```python
dataset.head(3)
```

```python
sns.distplot(dataset['ApplicantIncome_ss'])
plt.show()
```

```python
# now we compare both the graph 
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.title('Before Scalling')
sns.distplot(dataset['ApplicantIncome'])
plt.subplot(1,2,2)
plt.title('After Scalling')
sns.distplot(dataset['ApplicantIncome_ss'])
plt.show()
```
