## Z-Score Method to find Outliers

```python
# load library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import dataset
dataset = pd.read_csv('loan.csv')
dataset.head()
```

```python
z_score = (dataset['CoapplicantIncome'] - dataset['CoapplicantIncome'].mean())/(dataset['CoapplicantIncome'].std())
```

```python
z_score
```

```python
# Now put this data in to original dataset
dataset['Z_Score'] = z_score
```

```python
dataset
```

```python
# Now remove outliers (3 Z Score se less value ) 
dataset[dataset['Z_Score']<3]
```
