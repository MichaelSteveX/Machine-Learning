## Outlier Remove usning IQR Method (Inter Quantile Range)

```python
# so first check the shape of data set 
dataset.shape
```

```python
 # Now find the Quartile Q1 
q1 = dataset['CoapplicantIncome'].quantile(0.25)
q1
```

```python
 # Now find the Quartile Q2
q3 = dataset['CoapplicantIncome'].quantile(0.75)
q3
```

```python
IQR = q3-q1
IQR
```

```python
min_range = q1-(1.5 * IQR)
max_range = q3+(1.5 * IQR)
min_range,max_range
```

```python
# Now remove outliers from the original dataset
# first call the dataset 
dataset
```

```python
 #Here we call the data which is less to the maximum range 
new_data = dataset[dataset['CoapplicantIncome']<=max_range]
new_data
```
