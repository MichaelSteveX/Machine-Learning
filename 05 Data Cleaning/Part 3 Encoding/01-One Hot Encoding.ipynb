```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# import data
dataset = pd.read_csv('Loan.csv')
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
# Now fill the missing values in gender  
dataset['Gender'].fillna(dataset['Gender'].mode()[0],inplace = True)
```

```python
# Now fill the missing values in male 
dataset['Married'].fillna(dataset['Married'].mode()[0],inplace = True)
```

```python
dataset.isnull().sum()
```

```python
# extract data jisme humko encoding karna hai 
en_data = dataset[['Gender','Married']]
en_data
```

### get_dummies

```python
dummies = pd.get_dummies(en_data).head()
dummies.info()
```

### One Hot Encoding 

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first')
en = encoder.fit_transform(en_data).toarray()
en
```

```python
data1 = pd.DataFrame(en,columns = ['Gender_Male','Married_Yes'])
data1
```
