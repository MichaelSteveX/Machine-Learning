# Impute missing value with (SCIKIT LEARN)

```python
# import libraray
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# load data set
dataset = pd.read_csv('loan.csv')
dataset.head()
```

```python
# Now extract numerical column 
dataset.select_dtypes(include='float64').columns
```

#### Fill missing value using the SCIKIT LEARN 

```python
# import scikit learn library
from sklearn.impute import SimpleImputer
```

### Mean

```python
si = SimpleImputer(strategy='mean')
ar = si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

```python
# now convert the mean value in to dataframe 
new_dataset = pd.DataFrame(ar,columns =dataset.select_dtypes(include='float64').columns)
new_dataset 
```

### Median

```python
si = SimpleImputer(strategy='median')
md = si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

```python
new_data1 = pd.DataFrame(md,columns =dataset.select_dtypes(include='float64').columns )
new_data1
```

### Mode

```python
# retrive the categorical column 
dataset.select_dtypes(include='object').columns
```

```python
# Now apply function 
si = SimpleImputer(strategy='most_frequent')
mod = si.fit_transform(dataset[['Loan_ID', 'Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'Property_Area', 'Loan_Status']])
```

```python
# Now convert in to data frame 
new_data2 = pd.DataFrame(mod,columns = dataset.select_dtypes(include='object').columns)
new_data2
```
