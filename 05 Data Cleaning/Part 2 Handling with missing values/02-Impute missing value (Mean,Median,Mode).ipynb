# 02 Imputing Categorical data

```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data set
dataset = pd.read_csv('loan.csv')
dataset.head(5)
```

```python
# missing content
dataset.isnull().sum()
```

### Backward filling 

```python
dataset.fillna(method='bfill',axis = 0) # row wise filling 
dataset.fillna(method='bfill',axis = 1).head(3) # column wise filling 
```

### Forward filling 

```python
dataset.fillna(method='ffill',axis = 0)   # row wise filling
dataset.fillna(method='ffill',axis = 1).head(3)   # column wise filling 
```

## Mode Filling 

```python
# finding mode 
dataset['Gender'].mode()[0]
```

```python
# fill mode value in single column 
dataset['Gender'].fillna(dataset['Gender'].mode()[0],inplace=True)
```

```python
# now finnd all categorical column which have null value 
dataset.select_dtypes(include='object').columns
```

```python
dataset.select_dtypes(include='object').isnull().sum()
```

```python
# Now fill mode value in all the categorical column
for i in dataset.select_dtypes(include='object').columns:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

```python
# all categorical column fill with the mode and taht showing zero now the numerical column is remain 
dataset.isnull().sum()  
```

## Mean Filling

```python
# now fill the mean on numerical column 
dataset.select_dtypes(include = 'float64').columns
```

```python
dataset['LoanAmount'].mean()
```

```python
for i in dataset.select_dtypes(include = 'float64').columns:
    dataset[i].fillna(dataset[i].mean(),inplace = True)
```

```python
dataset.isnull().sum() 
```

## Median Filling

```python
dataset['LoanAmount'].median()
```

```python
for i in dataset.select_dtypes(include='float64'):
    dataset[i].fillna(dataset[i].median(),inplace = True)
```

```python
dataset.isnull().sum() 
```
